**Import Dataset**

```python
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Movies%20Recommendation.csv')
```

**Get Feature Selection**

```python
df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['Movie_Genre']+''+df_features['Movie_Keywords']+''+df_features['Movie_Tagline']+''+df_features['Movie_Cast']
```

```python
x
```

```python
x.shape
```

**get Features Text Conversion To tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

**Get similarity score**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

**Get Movie Name as Input from User and Validate for Closest Spelling**

```python
Favourite_Movie_Name = input('Enter your favourite movie name :')
```

```python
All_Movies_title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_title_List)
```

```python
print(Movie_Recommendation)
```

```python
Close_Match = Movie_Recommendation[0]
Close_Match
```

```python
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Index_of_Close_Match_Movie
```

```python
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
```

```python
Recommendation_Score
```

```python
len(Recommendation_Score)
```

**Get all movies sort based on recommendation score wrt favourite movies**

```python

Sorted_Similar_Movie = sorted(Recommendation_Score,key=lambda x:x[1],reverse = True)
Sorted_Similar_Movie
```

```python
i=1
for movie in Sorted_Similar_Movie:
  index = movie[0]
  title_from_index = df[df.index == index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```

```python
Favourite_Movie_Name = input('Enter your favourite movie name :')
All_Movies_title_List = df['Movie_Title'].tolist()
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_title_List)
Close_Match = Movie_Recommendation[0]
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
Sorted_Similar_Movie = sorted(Recommendation_Score,key=lambda x:x[1],reverse = True)
i=1
for movie in Sorted_Similar_Movie:
  index = movie[0]
  title_from_index = df[df.index == index]['Movie_Title'].values[0]
  if(i<11):
    print(i,'.',title_from_index)
    i+=1
```
